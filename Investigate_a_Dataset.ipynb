```python
# converting some columns that has date to a datetime datatype

df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
```

```python
#check for duplicate

df.duplicated().sum()
```

```python

    df.rename(columns={'No-show':'No_show'}, inplace='true')
```

```python
#on no show data converting the yes to 1 and no to 0 for easy manipulation 

df.No_show[df['No_show'] == 'Yes'] = '1'
df.No_show[df['No_show'] == 'No'] = '0'
df['No_show'] = pd.to_numeric(df['No_show'])
```

```python
#creating diffrent colomns for th

showed = df['No_show'] == 0
not_showed = df['No_show'] == 1
df['showed'] = showed
df['not_showed'] = not_showed
```

```python
#checking on both the oldest and youngest person

print('minimum age:',df['Age'].min()) 

print('maximum age:',df['Age'].max()) 
```

```python
#since no one is -1 years old drop -1 vavue mem

df=  df.drop(df[df["Age"]==-1].index)

df.Age.min()
```

### Research Question 1 (Gender committed to attending appointment?)

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.
#which gender attands more appointments

No_Showt_gender = df.groupby('Gender')['No_show'].value_counts().unstack()
No_Showt_gender
```

```python
#relationship btw gender and appointment

plot1 =df.groupby('Gender')['No_show'].value_counts().unstack().plot.bar()
plt.title("relation between Gender and appointments")
xlabel=('Gender')
ylabel=('appointments')
```

#Summary
